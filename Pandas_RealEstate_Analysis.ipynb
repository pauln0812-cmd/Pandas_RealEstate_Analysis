```python
df_concat = pd.concat([df1,df2,df3],ignore_index=True)
```

```python
file=df_concat.to_csv("file_cleaned1.csv")
```

```python
fig = px.scatter_mapbox(df_concat,lat="lat",lon="lon",hover_name="state",hover_data=["price_usd"],zoom=3,height=300,size=df_concat["price_usd"])
fig.update_layout(
    mapbox_style="white-bg",
    mapbox_layers=[
        {
            "below": 'traces',
            "sourcetype": "raster",
            "sourceattribution": "United States Geological Survey",
            "source": [
                "https://basemap.nationalmap.gov/arcgis/rest/services/USGSImageryOnly/MapServer/tile/{z}/{y}/{x}"
            ]
        }
      ])
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
fig=px.scatter(df_concat,y="price_usd",x="area_m2",trendline="ols",color="property_type")
fig.show()
```

```python
state_avg = df_concat.groupby("state").agg({"price_usd":"mean"}).sort_values(by="price_usd",ascending=False)
```

```python
state_avg["state"]=state_avg.index
```

```python
state_avg
```

```python
fig =px.bar(state_avg,x="state",y="price_usd")
fig.update_layout(title={'text':'Average House Price Per State'})
fig.show()
```
